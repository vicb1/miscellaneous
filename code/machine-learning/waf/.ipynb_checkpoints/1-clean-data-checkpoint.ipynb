```python
import numpy as np
import pandas as pd

df_dcop = pd.read_csv('C:/_WaferMLdata/test-short/df_dcop.csv', converters={'lotid': str})

df_dcop = df_dcop[df_dcop['compids'] != ' ']
df_dcop = df_dcop.sort_values(by=['compids', 'testopno', 'itemprompt', 'revdate'], ascending=[False, False, False, False])
df_dcop['unnecessary?'] = df_dcop.duplicated(subset=['compids', 'testopno', 'itemprompt'], keep='first')
df_dcop = df_dcop[df_dcop['unnecessary?'] == False]

# df_dcop.columns = df_dcop.columns.get_level_values(0)

df_dcop['testopno_itemprompt_char'] = df_dcop.apply(lambda row: row['testopno'] + '_' + row['itemprompt'] + '_char', axis=1)
df_dcop['testopno_itemprompt_num'] = df_dcop.apply(lambda row: row['testopno'] + '_' + row['itemprompt'] + '_num', axis=1)
df_dcop['testopno_itemprompt_date'] = df_dcop.apply(lambda row: row['testopno'] + '_' + row['itemprompt'] + '_date', axis=1)

df_dcop = df_dcop.reset_index()

df_dcop.describe(include='all')
```

```python
# df_dcop = df_dcop[:10000]
df_dcop_piv_char = pd.pivot_table(df_dcop, values='charactervalue', 
                             index=['compids'], 
                             columns=['testopno_itemprompt_char'], 
                             aggfunc='first')
df_dcop_piv_char = df_dcop_piv_char.reset_index(['compids'])
df_dcop_piv_num = pd.pivot_table(df_dcop, values='numericvalue', 
                             index=['compids'], 
                             columns=['testopno_itemprompt_num'], 
                             aggfunc='first')
df_dcop_piv_num = df_dcop_piv_num.reset_index(['compids'])

df_dcop_piv_date = pd.pivot_table(df_dcop, values='datevalue', 
                             index=['compids'], 
                             columns=['testopno_itemprompt_date'], 
                             aggfunc='first')
df_dcop_piv_date = df_dcop_piv_date.reset_index(['compids'])

df_dcop_piv = pd.merge(df_dcop_piv_char, df_dcop_piv_num, how='left', on=['compids', 'compids'])
df_dcop_piv = pd.merge(df_dcop_piv, df_dcop_piv_date, how='left', on=['compids', 'compids'])

df_dcop_piv['compids'] = df_dcop_piv.apply(lambda row: row['compids'].strip().upper(), axis=1)
df_dcop_piv = df_dcop_piv.rename(columns = {'compids': 'WaferID'})
df_dcop_piv = df_dcop_piv.set_index('WaferID')

df_dcop_piv.index = df_dcop_piv.index.astype(str)

# df_dcop_piv.to_csv('C:/_WaferMLdata/df_dcop_piv2.csv', encoding='utf-8', index=True)

df_dcop_piv.describe(include='all')
```
